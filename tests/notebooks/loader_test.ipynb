```python
import matplotlib.pyplot as plt
import numpy as np
from dataset_loader import WasteDatasetLoader
import cv2 
```

```python
def test_dataset_loading():
    """Test and visualize the loaded datasets"""
    
    # Initialize the loader
    loader = WasteDatasetLoader()
    
    # Load datasets
    trashnet_images, trashnet_labels = loader.load_trashnet()
    coco_images, coco_labels = loader.load_coco()
    combined_images, combined_labels = loader.load_combined_datasets()
    
    # Print dataset statistics
    print("\nDataset Statistics:")
    print("-" * 50)
    print(f"TrashNet: {len(trashnet_images)} images")
    print(f"COCO: {len(coco_images)} images")
    print(f"Combined: {len(combined_images)} images")
    
    # Show class distribution
    categories = ['glass', 'paper', 'cardboard', 'plastic', 'metal', 'trash']
    
    plt.figure(figsize=(15, 5))
    
    # TrashNet distribution
    plt.subplot(131)
    plt.bar(categories, np.bincount(trashnet_labels, minlength=6))
    plt.title('TrashNet Class Distribution')
    plt.xticks(rotation=45)
    
    # COCO distribution
    plt.subplot(132)
    plt.bar(categories, np.bincount(coco_labels, minlength=6))
    plt.title('COCO Class Distribution')
    plt.xticks(rotation=45)
    
    # Combined distribution
    plt.subplot(133)
    plt.bar(categories, np.bincount(combined_labels, minlength=6))
    plt.title('Combined Class Distribution')
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
    
    # Visualize sample images
    plt.figure(figsize=(15, 5))
    for i in range(5):
        plt.subplot(1, 5, i+1)
        img_rgb = cv2.cvtColor(combined_images[i], cv2.COLOR_BGR2RGB)
        plt.imshow(img_rgb)
        plt.title(categories[combined_labels[i]])
        plt.axis('off')
    
    plt.suptitle('Sample Images from Combined Dataset')
    plt.show()
```

```python

test_dataset_loading()
```
